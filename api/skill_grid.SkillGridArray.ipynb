# skill_grid.SkillGridArray

``` python
skill_grid.SkillGridArray(data)
```

A SkillGridArray is a single metric-SkillGrid, corresponding to a
“column” in a SkillGrid

Typically created by indexing a SkillGrid object, e.g. `ss["bias"]`.

## Examples

``` python
>>> gs = cc.gridded_skill()
>>> gs["bias"].plot()
```

## Attributes

| Name | Description |
|------------------------------------|------------------------------------|
| [coords](#modelskill.skill_grid.SkillGridArray.coords) | Coordinates (same as xr.DataSet.coords) |
| [mod_names](#modelskill.skill_grid.SkillGridArray.mod_names) | List of model names |
| [obs_names](#modelskill.skill_grid.SkillGridArray.obs_names) | List of observation names |
| [x](#modelskill.skill_grid.SkillGridArray.x) | x-coordinate values |
| [y](#modelskill.skill_grid.SkillGridArray.y) | y-coordinate values |

## Methods

| Name | Description |
|------------------------------------|------------------------------------|
| [plot](#modelskill.skill_grid.SkillGridArray.plot) | wrapper for xArray DataArray plot function |

### plot

``` python
skill_grid.SkillGridArray.plot(**kwargs)
```

wrapper for xArray DataArray plot function

#### Parameters

| Name | Type | Description | Default |
|---------|-------|-----------------------------------------------|----------|
| \*\*kwargs | Any | keyword arguments passed to xr.DataArray plot() e.g. figsize | `{}` |

#### Examples

``` python
>>> gs = cc.gridded_skill()
>>> gs["bias"].plot()
>>> gs.sel(model='SW_1').rmse.plot()
>>> gs.r2.plot(cmap='YlOrRd', figsize=(10,10))
```
