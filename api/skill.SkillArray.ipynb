# 

# skill.SkillArray

``` python
skill.SkillArray(data)
```

Visualize skill metric.

Obtained by selecting a single metric from a SkillTable.

## Examples

``` python
>>> sk = cc.skill()   # SkillTable
>>> sk.rmse           # SkillArray
>>> sk.rmse.plot.line()
```

## Attributes

| Name | Description |
|------------------------------------|------------------------------------|
| [name](#modelskill.skill.SkillArray.name) | Name of the metric |
| [plot](#modelskill.skill.SkillArray.plot) | Plot using the SkillArrayPlotter |

## Methods

| Name | Description |
|------------------------------------|------------------------------------|
| [to_dataframe](#modelskill.skill.SkillArray.to_dataframe) | Convert SkillArray to pd.DataFrame |
| [to_geodataframe](#modelskill.skill.SkillArray.to_geodataframe) | Convert SkillArray to geopandas.GeoDataFrame |

### to_dataframe

``` python
skill.SkillArray.to_dataframe(drop_xy=True)
```

Convert SkillArray to pd.DataFrame

#### Parameters

| Name    | Type | Description                                 | Default |
|---------|------|---------------------------------------------|---------|
| drop_xy | bool | Drop the x, y coordinates?, by default True | `True`  |

#### Returns

| Name | Type         | Description                |
|------|--------------|----------------------------|
|      | pd.DataFrame | Skill data as pd.DataFrame |

### to_geodataframe

``` python
skill.SkillArray.to_geodataframe(crs='EPSG:4326')
```

Convert SkillArray to geopandas.GeoDataFrame

Note: requires geopandas to be installed

Note: requires x and y columns to be present

#### Parameters

| Name | Type | Description | Default |
|------|------|-----------------------------------------------------|---------|
| crs | str | Coordinate reference system identifier passed to the GeoDataFrame constructor, by default “EPSG:4326” | `'EPSG:4326'` |

#### Returns

| Name | Type             | Description                |
|------|------------------|----------------------------|
|      | gpd.GeoDataFrame | Skill data as GeoDataFrame |
