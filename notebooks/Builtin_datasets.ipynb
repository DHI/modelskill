# Built-in datasets for demonstration purposes

* Oresund water levels
* Vistula river discharge

```python
import modelskill as ms
```

## Oresund water level

Compare tide gauges to MIKE 21 HD results for the Oresund strait.

```python
cc = ms.data.oresund()
cc
```

```python
cc.skill(metrics=["rmse","mae","cc","r2"]).style()
```

```python
for cmp in cc:
    cmp.plot.timeseries(figsize=(8, 3))
```

## Vistula river discharge

Compare river discharge to DHI's Global Hydrological Model (GHM) for the Vistula river in Poland.

```python
cc = ms.data.vistula()
cc
```

```python
cc.skill(metrics=["kge", "nse"]).style(show_best=False)
```

```python
cc.skill(metrics=["kge"])["kge"].plot.barh(figsize=(5, 3));
```

```python
for cmp in cc:
    cmp.plot.timeseries(figsize=(8, 3))
```
