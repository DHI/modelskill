# Plotting timeseries

The focus for this notebook is to show options for plotting matched data as timeseries using the Comparer object. 

Observation and ModelResults can also be plotted as timeseries with similar syntax but this is not covered in this notebook. 

```python
import pandas as pd
import modelskill as ms
```

```python
df = pd.read_csv("../tests/testdata/matched_data.csv", index_col='datetime', parse_dates=True)
cmp = ms.from_matched(df.loc["2019-01-01"], obs_item='Station X')
```

Default time series plot 

```python
cmp.plot.timeseries();
```

Use the `color` argument to change the color of the models (and potentially also the observation)

```python
cmp.plot.timeseries(color=["#729456", "red", "k"]);
```

You can also change the style of lines using the `style` argument similar to the `style` argument of pandas `df.plot()`

```python
cmp.plot.timeseries(style=["--", "-", "k-."]);
```
